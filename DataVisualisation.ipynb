Representation of a raw data set in the form of
 1.Bar chart
 2.Pie chart
 3.Line graph

```python
#importing Module required for Data visualisation
import pandas as pd # primary data structure library
import numpy as np  # Useful for many scientific computing in python 
import matplotlib.pyplot as plt

df=pd.read_csv("C:\\Users\\HP\\Desktop\\ML\\project\\Topic_Survey_Assignment.csv") 

#Let's view the elements 
df.head(6)
```

```python
#Rename a column so that it make sense
df.rename(columns={'Unnamed: 0':'Data_science_topic'}, inplace=True)

#let's view the elements and see how the data frame is change
df.head(6)
```

```python
#sort the data frame in descending order by "Very interested" column
df.sort_values(['Very interested'],ascending=False,inplace=True)
data_science_topic=df.loc[0]

df.head(6)
```

```python
# Defining a percentage function
def per(part):
    whole=2233;
    return 100*(part/whole)

# Forming a new Data Frame by percentage value of every data
data={'Data_science_topic':['Data Analysis / Statistics','Machine Learning','Data Visualization','Big Data (Spark / Hadoop)',
                            'Deep Learning','Data Journalism'],
      'Very interested':[per(1688),per(1629),per(1340),per(1332),per(1263),per(429)],
      'Somewhat interested':[per(444),per(477),per(734),per(729),per(770),per(1081)],
      'Not interested':[per(60),per(74),per(102),per(127),per(136),per(610)]}
df_percentage=pd.DataFrame(data)

# Let's view the new Data Frame
df_percentage.head(6)
```

Representation of above details in a bar chart

Ploting the new Data Frame as a Bar chart. 

```python
objects=('Data Analysis / Statistics','Machine Learning','Data Visualization','Big Data (Spark / Hadoop)','Deep Learning','Data Journalism')

df_percentage.plot(kind='bar',
                   legend=True,
                   figsize=(20,6),           
                   color=['#5cb85c','#5bc0de','#d9534f'],  # add custom color
                   width=0.8)                  # add width

arr=df_percentage['Very interested'].value_counts()
y_pos=np.arange(len(arr))
plt.xticks(y_pos,objects)        
plt.title('Percentage of responders interested in different data science areas') # add title to the plot
plt.ylabel('Percentage of responders') # add y-label to tha plot

plt.show()
```

Ploting the Data Frame as a Pie chart. 

```python
# Pie chart of very interested responders of differt data science topic
colors_list = ['gold', 'yellowgreen', 'lightcoral', 'lightskyblue', 'lightgreen', 'pink']

df['Very interested'].plot(kind='pie',
                           figsize=(15,6),
                           autopct='%1.1f%%', #add in percentages
                           startangle=90,     #start angle 90
                           explode=[0.1,0,0,0,0,0], #explode the largest value
                           shadow=True,       #add shadow
                           colors=colors_list,#add custom color
                           labels=None)       #turn off labels on pie chart

# scale the title up by 12% to match pctdistance
plt.title('very interested responders of different data science topic',y=1.12)
plt.axis('equal')                                    #sets the pie charts to look like a circle
plt.legend(labels=df['Data_science_topic'],loc='upper right')   # add legend

plt.show()
```

```python
# Pie chart of somewhat interested responders of differt data science topic
colors_list = ['gold', 'yellowgreen', 'lightcoral', 'lightskyblue', 'lightgreen', 'pink']

df['Somewhat interested'].plot(kind='pie',
                           figsize=(15,6), 
                           autopct='%1.1f%%',    #add in percentage
                           startangle=90,        #start angle 90
                           explode=[0,0,0,0,0,0.1],  # explode the largest value
                           shadow=True,          # add shadow
                           colors=colors_list,   # add custom color
                           labels=None)          # turn off labels of the pie chart

# scale the title up by 12% to match pctdistance
plt.title('Somewhat interested responders of different data science topic',y=1.12)
plt.axis('equal')                                        #sets the pie charts to look like a circle
plt.legend(labels=df['Data_science_topic'],loc='upper right')   #add legend 

plt.show()
```

```python
# Pie chart of not interested responders of differt data science topic
colors_list = ['gold', 'yellowgreen', 'lightcoral', 'lightskyblue', 'lightgreen', 'pink']
df['Not interested'].plot(kind='pie',
                           figsize=(15,6),
                           autopct='%1.1f%%',    # add in percentage
                           startangle=90,         #start angle 90
                           explode=[0.1,0.1,0.1,0,0,0],   #explode small 3 values
                           shadow=True,         #add shadow
                           colors=colors_list,  #add custom color
                           labels=None)         # turn off labels off the pie chart

# scale the title up by 12% to match pctdistance
plt.title('Not interested responders of different data science topic',y=1.12)
plt.axis('equal')                               #sets the pie charts to look like a circle
plt.legend(labels=df['Data_science_topic'],loc='upper right')   # add legend

plt.show()
```

Ploting the new data Frame as a Line Graph

```python
objects=('Data Analysis / Statistics','Machine Learning','Data Visualization','Big Data (Spark / Hadoop)','Deep Learning','Data Journalism')
df_percentage.plot(kind='line',
                   figsize=(15,6))

arr=df_percentage['Very interested'].value_counts()
y_pos=np.arange(len(arr))
plt.xticks(y_pos,objects)

plt.title('Percentage of responders of different data science areas')  # add title to the line Graph
plt.ylabel('Percentage of responders')                  #add y-axis label

plt.show()
```
